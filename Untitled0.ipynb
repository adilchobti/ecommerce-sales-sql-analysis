```python
# === Colab: SQL -> results -> graphs (SQLite) ===
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
from google.colab import files
import os

# 1) Upload Excel
uploaded = files.upload()
xlsx_name = list(uploaded.keys())[0]
print("Loaded:", xlsx_name)

# 2) Read Excel (Online Retail II often has 2 sheets)
xls = pd.ExcelFile(xlsx_name)
print("Sheets:", xls.sheet_names)

dfs = []
for sh in xls.sheet_names:
    df_sh = pd.read_excel(xlsx_name, sheet_name=sh)
    # Normalize column names (strip spaces)
    df_sh.columns = [c.strip() for c in df_sh.columns]
    dfs.append(df_sh)

df = pd.concat(dfs, ignore_index=True)

# 3) Basic cleaning / typing (keep it simple & robust)
# Expected columns: Invoice, StockCode, Description, Quantity, InvoiceDate, Price, Customer ID, Country
# Some files use slightly different names; handle common variants.
col_map = {}
for c in df.columns:
    c_low = c.lower()
    if c_low in ["invoice", "invoiceno", "invoice no", "invoice_no"]:
        col_map[c] = "Invoice"
    elif c_low in ["stockcode", "stock code", "stock_code"]:
        col_map[c] = "StockCode"
    elif c_low in ["description"]:
        col_map[c] = "Description"
    elif c_low in ["quantity"]:
        col_map[c] = "Quantity"
    elif c_low in ["invoicedate", "invoice date", "invoice_date"]:
        col_map[c] = "InvoiceDate"
    elif c_low in ["price", "unitprice", "unit price", "unit_price"]:
        col_map[c] = "Price"
    elif c_low in ["customer id", "customerid", "customer_id"]:
        col_map[c] = "CustomerID"
    elif c_low in ["country"]:
        col_map[c] = "Country"

df = df.rename(columns=col_map)

required = ["Invoice", "StockCode", "Quantity", "InvoiceDate", "Price", "CustomerID", "Country"]
missing = [c for c in required if c not in df.columns]
if missing:
    raise ValueError(f"Missing expected columns: {missing}. Found: {list(df.columns)}")

df["Invoice"] = df["Invoice"].astype(str).str.strip()
df["StockCode"] = df["StockCode"].astype(str).str.strip()
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["CustomerID"] = pd.to_numeric(df["CustomerID"], errors="coerce")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

# Common cleaning: remove nulls, returns/cancellations, negative/zero qty or price
df = df.dropna(subset=["InvoiceDate", "Quantity", "Price", "CustomerID"])
df = df[(df["Quantity"] > 0) & (df["Price"] > 0)]
# Cancelled invoices often start with "C"
df = df[~df["Invoice"].str.upper().str.startswith("C")].copy()

# Add revenue column
df["Revenue"] = df["Quantity"] * df["Price"]

print("Clean rows:", df.shape[0])

# 4) Load into SQLite (single table is enough for SQL analytics)
conn = sqlite3.connect(":memory:")
df.to_sql("transactions", conn, index=False, if_exists="replace")

# Helpful indexes (optional)
conn.execute("CREATE INDEX IF NOT EXISTS idx_date ON transactions(InvoiceDate);")
conn.execute("CREATE INDEX IF NOT EXISTS idx_invoice ON transactions(Invoice);")
conn.execute("CREATE INDEX IF NOT EXISTS idx_customer ON transactions(CustomerID);")
conn.execute("CREATE INDEX IF NOT EXISTS idx_stock ON transactions(StockCode);")

# 5) SQL QUERIES (produce data for 3 graphs)

# (A) Monthly revenue
sql_monthly_revenue = """
WITH base AS (
  SELECT
    strftime('%Y-%m', InvoiceDate) AS year_month,
    SUM(Revenue) AS revenue
  FROM transactions
  GROUP BY 1
)
SELECT year_month, revenue
FROM base
ORDER BY year_month;
"""

# (B) Top 10 products by revenue
sql_top_products = """
SELECT
  StockCode,
  COALESCE(NULLIF(TRIM(Description), ''), '(No description)') AS Description,
  SUM(Revenue) AS revenue
FROM transactions
GROUP BY 1, 2
ORDER BY revenue DESC
LIMIT 10;
"""

# (C) Customer segmentation by number of orders (RFM-lite, simple & visual)
# Segments:
# - Occasional: 1 order
# - Regular: 2-5 orders
# - Premium: 6+ orders
sql_customer_segments = """
WITH orders_per_customer AS (
  SELECT
    CustomerID,
    COUNT(DISTINCT Invoice) AS nb_orders
  FROM transactions
  GROUP BY CustomerID
),
segmented AS (
  SELECT
    CASE
      WHEN nb_orders = 1 THEN 'Occasional'
      WHEN nb_orders BETWEEN 2 AND 5 THEN 'Regular'
      ELSE 'Premium'
    END AS segment
  FROM orders_per_customer
)
SELECT
  segment,
  COUNT(*) AS customers
FROM segmented
GROUP BY segment
ORDER BY
  CASE segment
    WHEN 'Occasional' THEN 1
    WHEN 'Regular' THEN 2
    ELSE 3
  END;
"""

# 6) Run SQL -> DataFrames
monthly = pd.read_sql_query(sql_monthly_revenue, conn)
top_products = pd.read_sql_query(sql_top_products, conn)
segments = pd.read_sql_query(sql_customer_segments, conn)

# 7) Make output folder and plots
os.makedirs("outputs", exist_ok=True)

# --- Graph 1: Monthly revenue (line) ---
plt.figure(figsize=(10,5))
plt.plot(monthly["year_month"], monthly["revenue"])
plt.xlabel("Month (Mois)")
plt.ylabel("Revenue (Chiffre d'affaires)")
plt.title("Monthly Revenue (CA mensuel)")
plt.grid(True)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("outputs/revenue_monthly.png", dpi=200)
plt.show()

# --- Graph 2: Top 10 products by revenue (barh) ---
tp = top_products.copy()
tp["label"] = tp["StockCode"].astype(str) + "  " + tp["Description"].astype(str)
tp = tp.sort_values("revenue", ascending=True)

plt.figure(figsize=(10,6))
plt.barh(tp["label"], tp["revenue"])
plt.xlabel("Revenue (Chiffre d'affaires)")
plt.ylabel("Product (Produit)")
plt.title("Top 10 Products by Revenue (Top 10 produits par CA)")
plt.grid(True, axis="x")
plt.tight_layout()
plt.savefig("outputs/top_products.png", dpi=200)
plt.show()

# --- Graph 3: Customer segments (bar) ---
plt.figure(figsize=(8,5))
plt.bar(segments["segment"], segments["customers"])
plt.xlabel("Segment (Segment)")
plt.ylabel("Number of customers (Nombre de clients)")
plt.title("Customer Segments by Number of Orders (Segmentation clients)")
plt.grid(True, axis="y")
plt.tight_layout()
plt.savefig("outputs/customer_segments.png", dpi=200)
plt.show()

print("Saved PNGs in outputs/:")
print("- outputs/revenue_monthly.png")
print("- outputs/top_products.png")
print("- outputs/customer_segments.png")
```
